## QUESTION 1.

**Remainder Grouping Dictionary**

Write a Python program that reads two inputs from the user:

A single line containing a list of integers separated by commas. The line may contain arbitrary spaces around the commas.
An integer k - the divisor.
Your task is to build a dictionary that groups the numbers by their remainder when divided by k.

1. Each key is a remainder r (0<=r<k) that actually appears among the given numbers.
2. The value for a key r is a list of all numbers whose remainder equals r, sorted in ascending order.

Print the dictionary in ascending order of the keys using the following format (one key per line) with values in ascending order.:

key1 - val1,val2,...

key2 - val1,val2,...

...

There must be no extra spaces around the dash or the commas.

````python
input_line = input()
numbers = [int(i.strip()) for i in input_line.split(",")] # read a list of integers from input, separated by commas 
k = int(input())

d={}
for n in numbers:
    r = n%k # calculate remainder of n when divided by k
    if r in d:
        d[r].append(n) # append number n to the list for remainder r
    else:
        d[r] = [n] # create a new list for remainder r with number n as the first element
print(f"dictonary d is: {d}")        

for remainder in sorted(d): # sort the remainders in ascending order, example: {1: [1, 19, 7], 4: [10, 22], 5: [5]} -> [1,4,5]
    numbers_list = d[remainder] # get the list of numbers for this remainder, example: [10,22]
    numbers_list.sort() # sort the list of numbers for this remainder, example: [4,1,7] -> [1,4,7]
    
    value_str = ",".join(map(str, numbers_list)) # convert the sorted list of numbers to a comma-separated string
    
    print(f"{remainder} - {value_str}") # print the remainder and the corresponding sorted list of numbers

# Group numbers by their remainders when divided by k and sort them.
# Read a list of integers and an integer k from input.
# Print the numbers grouped by their remainders when divided by k, sorted within each group
````

**Prime Galore Count**

**Task:**

Write a Python function called `prime_galore(L)` that takes a list of non-negative integers $L$ as input.

The function must implement the following logic:

1.  It should iterate through the indices ($i$) of the list $L$.
2.  For each index $i$, it must check two conditions:
      * **Condition 1 (Index Prime):** The index $i$ itself must be a prime number.
      * **Condition 2 (Value Prime):** The value at that index, $L[i]$, must also be a prime number.
3.  The function should return the total count of elements in the list that satisfy *both* the Index Prime and Value Prime conditions.

You should use a helper function, `isprime(num)`, to efficiently determine if a given integer is a prime number.

**Example Run**

Given the list:

```python
L = [1, 3, 11, 18, 17, 23, 6, 8, 10]
```

The expected output is **2**.

| Index $i$ | $L[i]$ | $i$ is Prime? | $L[i]$ is Prime? | Both True? |
| :---: | :---: | :---: | :---: | :---: |
| 0 | 1 | False | False | False |
| 1 | 3 | False | True | False |
| **2** | **11** | **True** | **True** | **True** |

| 5 | 23 | True | True | False |

**Note on Prime Numbers**

  * A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
  * The first few prime numbers are 2, 3, 5, 7, 11, 13, and so on.

````python
def isprime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True
# prime index, value at prime index is also prime

def prime_galore(L):
    count = 0
    for i in range(len(L)):
        if isprime(i) and isprime(L[i]):
            count += 1
    return count
# driver code
L = [1,3,11,18,17,23,6,8,10]
print(prime_galore(L))
````

**Print Pieces Moved from Chess Notation string.**

Write a program that reads chess moves in standard algebraic notation and prints out the names of the chess pieces that have been moved. For castling moves, both the king and rook should be printed.

NOTE: This is an I/O type question. You need to write the complete code to read the input and print the output.

**Input Format**

The input will be a string of chess moves in format "{move_number} {white_move} {black_move}, where each move is given in standard algebraic notation.

**Output Format**

Print the names of the pieces moved in each move. The output should be each piece's name in separate lines. For castling, print "King" and "Rook".
Chess Notation Details
* Pieces are represented by their initials:
K - King

R - Rook

N - Knight
Pawn (no initial, just represented by move)

* Castling is represented with:
O-O (kingside castling)

**Example:**

    1. e4 e5 2. Nf3 Nc6 3. Bb5 a6 4. O-O Qc7

````python
def print_pieces_moved(moves_string):
    """
    Reads chess moves in standard algebraic notation and prints the names of the 
    pieces that have been moved.
    """
    
    # Standard piece mapping for single letter initials
    piece_map = {
        'K': "King",
        'Q': "Queen",
        'R': "Rook",
        'B': "Bishop",
        'N': "Knight"
    }
    
    # Split the input string into individual move components (including move numbers)
    # The split() function handles arbitrary whitespace between moves.
    parts = moves_string.split()
    
    # Iterate through the components, ignoring the move numbers
    for move in parts:
        
        # 1. Skip move numbers (e.g., "1.", "2.", "3.")
        if move.endswith('.'):
            continue
            
        # 2. Check for Castling first, as it's a special case (King and Rook move)
        if move == "O-O" or move == "0-0":
            # Kingside Castling
            print("King")
            print("Rook")
            continue
            
        if move == "O-O-O" or move == "0-0-0":
            # Queenside Castling
            print("King")
            print("Rook")
            continue
            
        # 3. Check for the piece initial
        # The first character of the move will be the piece initial 
        # if it's not a pawn move.
        piece_initial = move[0]
        
        if piece_initial in piece_map:
            # The move is for King, Queen, Rook, Bishop, or Knight
            print(piece_map[piece_initial])
        else:
            # If the first character is not a piece initial, it must be 
            # a file (a-h), meaning it's a Pawn move (e.g., e4, exd4).
            print("Pawn")

# --- Driver Code (Simplified Input Reading) ---
try:
    # Read the entire line of input using the standard input() function
    input_line = input().strip()
    
    # Process the moves
    if input_line:
        print_pieces_moved(input_line)

except EOFError:
    # This block handles the case where no input is available
    pass
except Exception:
    # Catch any other runtime issues, like unexpected input formats
    pass
````

**Approach and Logic**

**Note**

````python
''' Approach 1: Using list of characters to modify specific indices '''

# Read the number of items
r = int(input())

# Read the space-separated list of indices
# This handles any number of indices, as required by the problem.
# Example: "0 2 5 1" -> [0, 2, 5, 1]
indexes_to_mark = list(map(int, input().split()))

# Reading tasks as list of characters
# Note: The input format uses a space ' ' inside the brackets: - [ ]
tasks = [list(input()) for _ in range(r)]

def mark(i):
    # Checking valid index (0-based)
    if 0 <= i < r:
        # Based on the input format, the ' ' or 'x' is always at index 3
        # E.g., '-' is 0, ' ' is 1, '[' is 2, ' ' is 3, ']' is 4...
        tasks[i][3] = 'x'

# Iterate over all collected indices
for index in indexes_to_mark:
    mark(index)

# Print the updated tasks
for t in tasks:
    print("".join(t))
````

## QUESTION 6.

**Sum Numbers Inside Square Brackets**

Given a multiline text, extract all numbers that appear inside square brackets in the whole text, compute their sum, and output this sum.

* There will be only valid integers inside the square brackets (including negative).
* There will be no nested brackets and there will be always closing bracket in the same line.
* The brackets might span over multiple line(beginning in one line and ending in another).
* There may be other characters (including other brackets) outside the numeric brackets – ignore them.

**Input Format**

n

line_1

line_2

...

line_L

* The first line contains a single integer n – the number of following lines.
* Each of the next n lines is an arbitrary string that may contain zero or more integer numbers enclosed in square brackets [ and ].
* Output Format A single integer – the sum of all numbers found inside [ ].

**Example**

`Input`

4

The price is [12] dollars.

No brackets here.

Multiple [3][4][5] in one line.

[100] is the final one.

`Output`

124

**Explanation**

Numbers extracted: 12, 3, 4, 5, 100 → sum = 124.

````python
n = int(input())    # number of input lines
all_text = ""    # concatenate all input lines into a single string
for _ in range(n):    # read n lines
    all_text += input()    # concatenate each line to all_text

total_sum = 0    # Initialize sum of numbers
collecting_number = False    # Flag to indicate if we are inside brackets
current_number_str = ""    # To accumulate digits of the current number

for char in all_text:    # Iterate through each character
    if char == '[':    # Start of a number
        collecting_number = True    # Start collecting number
        current_number_str = ""    # Reset for new number
    elif char == ']' and collecting_number:    # End of a number 
        if current_number_str:    # Ensure there's something to convert
            total_sum += int(current_number_str) 
        current_number_str = ""    # Reset for next number
        collecting_number = False    # Stop collecting number
    elif collecting_number:    # Inside brackets
        is_digit = char.isdigit()    # Check for digits 0-9
        is_minus_sign = (char == '-') and (len(current_number_str) == 0)    # Allow '-' only at the start
        
        if is_digit or is_minus_sign:    # Valid character for number
            current_number_str += char    # Accumulate digit or minus sign
print(total_sum)    # Print the final sum of numbers found within brackets
````

>**Code Summary: Bracketed Integer Summation**

This program reads a specified number ($N$) of input lines and concatenates them into a single string (`all_text`). It then **parses** this entire string to find and sum all valid integers (positive or negative) that are strictly enclosed within square brackets (`[ ]`).

The core of the code is a **finite state machine** implemented using the boolean flag $\text{\texttt{collecting\_number}}$ to determine whether the parser is currently inside or outside a number sequence defined by the brackets.

>**Game Plan**

1.  **Read $N$ lines** and join them into one continuous string.
2.  **Iterate** through the string character by character.
3.  **State 1 (Outside):** Wait for $\text{\texttt{[}}$.
4.  **State 2 (Inside):** Collect digits and an initial minus sign (`-`) into $\text{\texttt{current\_number\_str}}$.
5.  **State Transition:** When $\text{\texttt{]}}$ is found, convert the collected string to an integer, add it to $\text{\texttt{total\_sum}}$, and switch back to **State 1**.

>**Tricky Part and Approach**

The tricky part of this problem is not the final sum, but correctly managing the **State** and the **Edge Cases** of the number format.

>**Tricky Part: The Negative Sign Edge Case**

The most complex condition is ensuring the minus sign (`-`) is only accepted if it's the very first character of the number string, preventing invalid inputs like `[1-0]` or `[10-]` from being incorrectly processed.

* **Approach:** The code elegantly handles this with two checks combined with the $\text{\texttt{collecting\_number}}$ flag:
    * $\text{\texttt{is\_digit = char.isdigit()}}$
    * $\text{\texttt{is\_minus\_sign = (char == '-') and (len(current\_number\_str) == 0)}}$

    This conditional logic ensures that only `[...123...]`, `[-123...]`, and digits following a minus sign are collected, while invalid characters or misplaced minus signs are ignored.

## QUESTION 7.

**Abbreviate Initials And Sort**

Write a Python function that takes a list of full names, processes them to create abbreviated forms in the format "Last, F.M." (where F and M are the initials of the first and middle names), and returns the list sorted alphabetically.

Assume that the names are given in the correct case.

Hint: Use sorted function or list.sort to sort the list

NOTE: This is an I/O type question, you need to write the whole code for taking input and printing the output.


**Input Format**

* First line contains the number of names, n.
* Next n lines contain one full name per line.

**Output Format**

* Output the processed names in sorted order, one per line.


**Example**

**Input:**

3

John Doe

Alice Johnson

Bob Alan Rickman

**Output:**


Doe, J.

Johnson, A.

Rickman, B.A.

````python
n = int(input())
ans = []
for _ in range(n):
    names = input().split()
    
    # 1. Collect only the initials (e.g., ['B.', 'A.'])
    initials_list = []
    for i in names[:-1]: # all names except the last one
        initials_list.append(i[0] + ".")
        
    # 2. Join the initials with NO space (e.g., "B.A.")
    initials_combined = "".join(initials_list) # joining without space
    
    # 3. Create the final required format: "Last, F.M."
    full_abbreviation = names[-1] + ", " + initials_combined
    
    # Append the correctly formatted string
    ans.append(full_abbreviation) 

ans.sort() # Sort the final list alphabetically    

for name in ans:
    print(name)
````

## QUESTION 8.

**Hand Cricket Match Runs**

Given two columns of numbers from 0 to 6, separated by space, played by batsman and bowler in a hand cricket match, find `the total number of balls played` and the `total runs scored by the batsman` at the end of the innings.

The match ends when the batsman plays the same number as the bowler or plays the same number twice continously.

The number batsman plays will be added to the runs, except when batsman plays 0, the number bowler plays will be added to the runs.

The number played when the game ends is not counted towards the runs.

**Input Format**

Each line with numbers played by batsman and bowler separated by space.
Output Format

Number of balls played and the total runs separated by space.

**Example**

`Input:`

1 4

2 3

6 2

5 1

0 2

3 3

`Output:`

6 16

**Explanation**


1 4 - 1

2 3 - 2

6 2 - 6

5 1 - 5

0 2 - 2 (0 played by batsman so 2 is counted as runs)

3 3 - out

Total: 1+2+2+5+2 = 16

`Input:`

1 4

5 3

0 3

5 1

5 4

`Output:`

5 14

**Explanation**

1 4 - 1

5 3 - 5

0 3 - 3 (0 played by batsman so 2 is counted as runs)

5 1 - 5

5 4 - out (batsman played 5 twice continously)

Total: 1+5+3+5 = 14

````python
# Initialize game state variables
total_runs = 0
balls_played = 0
# Initialize previous score outside the 0-6 range to ensure the first ball isn't an out.
previous_batsman_score = -1 

# The game runs in an infinite loop, relying on the 'break' statement 
# inside the loop when the 'out' condition is met.
while True:
    
    # We assume 'input()' will safely read the next line containing two numbers.
    # Read the line (e.g., "1 4")
    line = input()
    
    # If the input stream is exhausted (no more lines), we must assume the innings is over.
    # This line acts as a simple check for an empty line (EOF in some environments).
    if not line:
        break
        
    # Split the line and convert the two parts to integers
    R_B, R_Bo = map(int, line.split()) # R_B: Batsman's run, R_Bo: Bowler's run
    
    # --- Game Logic ---
    
    # 1. Update Balls Played: The ball is always played, even the last one.
    balls_played += 1
    
    # 2. Check for the OUT condition
    # The game ends if:
    # A) Batsman plays the same number as the Bowler (R_B == R_Bo)
    # B) Batsman plays the same number twice continuously (R_B == previous_batsman_score)
    if R_B == R_Bo or R_B == previous_batsman_score:
        # If out, the runs for this final ball are NOT counted.
        # We break the loop immediately.
        break
        
    # 3. Calculate Runs for the ball (Only executes if NOT out)
    if R_B > 0:
        # Batsman plays 1-6
        total_runs += R_B
    else: # R_B == 0
        # Batsman plays 0, Bowler's score is counted
        total_runs += R_Bo
    
    # 4. Update the previous batsman score for the next ball's continuous-out check
    previous_batsman_score = R_B
    
# Output the final result
print(f"{balls_played} {total_runs}")
````

**Hand Cricket Problem Summary**

The goal is to process a series of two-number inputs (Batsman's Move $R_B$ and Bowler's Move $R_{Bo}$) line by line until an "out" condition is met. You need to calculate the **total balls played** (including the one that resulted in the out) and the **total runs scored** (excluding the runs scored on the final out ball).

**Scoring Rules:**

  * **If $R_B > 0$:** Runs added are $R_B$.
  * **If $R_B = 0$:** Runs added are $R_{Bo}$.

**Out Conditions (Game Ends):**

1.  **Caught/Bowled:** $R_B = R_{Bo}$
2.  **Continuous:** $R_B$ is the same as the batsman's move on the immediately preceding ball.

**The Tricky Part and Approach**

The core difficulty lies in the **order of operations** inside the main `while` loop, specifically how you handle the ball count, the out check, and the run total.

**The Loop Structure and Order**

The critical sequence for every ball must be:

1.  **Read Input ($R_B, R_{Bo}$)**
2.  **Increment Ball Count:** The ball is now officially "played."
3.  **Check for Out Condition:** This must happen *before* calculating the runs.
4.  **If Out:** **BREAK** the loop immediately. This ensures the runs for this ball are never added.
5.  **If NOT Out:** Calculate and add the runs to the `total_runs`.
6.  **Update State:** Save the current $R_B$ to check for the continuous out condition on the *next* ball.

**Handling the Previous Score State**

To check for the continuous out condition (`R_B == previous_batsman_score`), you must initialize the `previous_batsman_score` to a value that *cannot* be played (e.g., -1 or 7).

  * **Initialization:** `previous_batsman_score = -1`
  * **Update:** The update must happen **at the end** of a non-out ball's processing: `previous_batsman_score = R_B`.

**The $0$ Run Rule**

The scoring rule for the batsman playing 0 is unusual. It's easy to miss that $R_B=0$ is **NOT** an out condition (unless $R_{Bo}$ is also 0, which triggers the Caught/Bowled condition).
