## QUESTION 1.

### **Column Totals in a Markdown Table (Numeric Columns Only)**

Read the contents of a file containing a markdown table with integer values and print the column sums.

**Markdown table format**

* First line contains the readers surrounded and separated by | character.
* Second line is header separater containing | and - and has no data.
* Every following line is a data row with values surrounded and separated by | character.
* Spaces may appear around the pipes and around the cell contents.
* All columns contain integer values (positive, negative or zero).

**Output format**

One column sum per line, in the format:

` <column_name>: <total> `

The order of the output must follow the order of the columns in the header.

**Note:** This is a fileinstdout type problem the whole input should be written to a temporary file first, and the solution must read from that file.

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================

import os

filename = 'file_handling_ques_1_input.txt'
table_content = """| Score |
|-------|
| 10    |
| 15    |
| 20    |"""

# ------------------------------------------
# SOLUTION: Write your code below this line
# ------------------------------------------

with open(filename, 'w') as f:
    f.write(table_content)

# Read the file and compute totals
try:
    with open(filename, "r") as f:
        # Filter out empty lines and strip whitespace
        lines = [line.strip() for line in f if line.strip()]

    if lines:
        # Parse Headers (removes leading/trailing pipes)
        headers = [h.strip() for h in lines[0].split('|') if h.strip()]
        num_columns = len(headers)
        column_sums = [0] * num_columns

        # Process data rows (skip header and separator line)
        for line in lines[2:]:
            cells = [c.strip() for c in line.split('|') if c.strip()]
            
            for i in range(min(len(cells), num_columns)):
                try:
                    column_sums[i] += int(cells[i])
                except ValueError:
                    pass # Ignore non-numeric data

        # 3. OUTPUT: Print in required format
        for i in range(num_columns):
            print(f"{headers[i]}: {column_sums[i]}")

except FileNotFoundError:
    print(f"Error: Could not find '{filename}'")
````

## QUESTION 2.

### **Uppercase Every k-th Vowel and lower case other vowels in a File**

Write a program that uppercases every k-th vowel in a text file and lowercase other vowels. Vowels are a, e, i, o, u and the check should be case-insensitive. The case of consonants should be unchanged.

Assume k is a positive integer.

* The first line of the file contains the integer k.
* Subsequent lines of the file contain the text.
* The vowel count is cumulative across the entire text. For example, if k=3, the 3rd, 6th, 9th, etc., vowels found in the file should be uppercased, regardless of which line they are on.

**NOTE:** This is a file-in-stdout type question where the input is read from the file and the output is printed in the standard output.

>Example

`Input`(contents of the file)

3

This is a sample text.

it has many vowels.

every third one becomes uppercase.

`Output`


This is A sample tExt.

it has mAny vowels.

Every third One becOmes uppErcase.

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'vowel_kth_input.txt'
test_data = """3
This is a sample text.
it has many vowels.
every third one becomes uppercase."""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Write your code below this line
# ------------------------------------------

def process_vowels(filename):
    try:
        with open(filename, 'r') as f:
            lines = f.readlines()
        
        if not lines:
            return

        # The first line is k
        k = int(lines[0].strip())
        
        # Subsequent lines are the text
        text_lines = lines[1:]
        
        vowels = "aeiouAEIOU"
        vowel_counter = 0
        
        for line in text_lines:
            result_line = ""
            for char in line:
                if char in vowels:
                    vowel_counter += 1
                    # Check if it's the k-th vowel
                    if vowel_counter % k == 0:
                        result_line += char.upper()
                    else:
                        result_line += char.lower()
                else:
                    # Consonants and symbols remain unchanged
                    result_line += char
            
            # Print the result for each line (end="" because line already has \n)
            print(result_line, end="")

    except FileNotFoundError:
        print(f"Error: {filename} not found.")
    except ValueError:
        print("Error: First line must be an integer k.")

process_vowels(filename)
````

* **Key Logic HighlightsModulo Operator (%):** By using `vowel_counter % k == 0`, we catch every $k, 2k, 3k \dots$ vowel.
* **The Global Counter:** Notice that vowel_counter is defined outside the line loop. This ensures that if the 2nd vowel is at the end of line 1, and $k=3$, the very first vowel of line 2 is correctly identified as the 3rd vowel.
* **Case Handling:** The problem asks to lowercase all vowels that are not the $k$-th one. Using .lower() on non-k-th vowels ensures compliance even if the input vowel was originally uppercase.
* **end="":** When printing the `result_line`, we use `end=""` because the original strings from `readlines()` already contain the newline character `\n`.

**Note:** The `vowel_counter` persists across lines. This demonstrates how to maintain state while reading a file line-by-line.

### **Total Size of Image Files**

Given a file containing file sizes and file names separated by a comma, find the total size of image files with extensions .jpg, .jpeg, .png, and .gif (case insensitive).

**Assumptions**

* The file is in the standard format where each line has a file size followed by a file name.
* The file size is in bytes.
* The file name is in the format filename.extension.

>Examples

`Input File`

2000,file1.jpg

890,file2.txt

30500,file3.JPEG

12000,file4.png

40000,file5.gif

490,file6.docx

`Output`

84500

**Explanation**

The total size of image files, which is 2000 + 30500 + 12000 + 40000 = 84500

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'ques_3_input.txt'
test_data = """
2000,file1.jpg
890,file2.txt
30500,file3.JPEG
12000,file4.png
40000,file5.gif
490,file6.docx
"""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Write your code below this line
# ------------------------------------------

img_extensions = {'jpg', 'jpeg', 'png', 'gif'}
with open(filename, 'r') as file:
    total_size = 0
    for line in file:
        parts = line.strip().split(',')
        if len(parts) < 2: continue # Skip empty or malformed lines
    
        size, filename = parts[0], parts[1]
        extension = filename.split('.')[-1].lower()
    
        if extension in img_extensions:
            total_size += int(size)
    print(total_size)
````

This code efficiently processes a text-based database to filter and aggregate information based on specific criteria. Here is a summary of the approach and the key technical points.

### **Approach Summary**

The solution follows a **linear processing** strategy:

1. **Filter Definition:** It establishes a reference set of valid image extensions (`.jpg`, `.jpeg`, `.png`, `.gif`).
2. **File Stream:** It opens the input file and iterates through it line by line to minimize memory usage (especially useful for large files).
3. **Parsing & Cleaning:** For each line, it extracts the file extension and the size using string manipulation.
4. **Validation & Accumulation:** It checks if the extracted extension matches the "image" criteria and, if so, adds the size to a running total.

---

### **Important Points & Key Features**

* **Case Insensitivity:** The code uses `.lower()` when checking extensions. This ensures that `file3.JPEG` is treated the same as `file1.jpg`, matching the requirements of the problem.
* **String Parsing Strategy:**
* `split(",")`: Separates the numeric size from the filename.
* `split(".")[-1]`: Targets the very last part of the filename to find the extension.
* `strip()`: Removes any hidden newline characters (`\n`) or whitespace at the end of the line that could interfere with the matching logic.


* **Data Type Conversion:** It explicitly converts the string representation of the size into an integer (`int()`) before performing addition.
* **Efficiency with Sets:** The use of `img_extensions = {...}` (a **Set**) provides  average time complexity for lookups, making the extension check very fast.

A file is provided where the first line contains a positive integer  representing the width of the zigzag. Each subsequent line contains exactly one character.

Your task is to print the characters in their original order, but prepend each character with a specific number of leading spaces based on a **zigzag pattern**:

1. Start with **0** spaces.
2. Increase the number of spaces by **1** per line until **** spaces are reached.
3. Decrease the number of spaces by **1** per line until **0** spaces are reached.
4. Repeat this up-and-down cycle for the remainder of the file.

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'zigzag_input.txt'
test_data = """3
a
p
p
l
e
o
r
a
n
g
e"""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Code
# ------------------------------------------

with open(filename, "r") as file:
    # Read the first line as the zigzag size
    line = file.readline().strip()
    if not line:
        z = 1 # Fallback if file is empty
    else:
        z = int(line)
    
    # Read the remaining characters into a list
    char_list = [line.strip() for line in file if line.strip()]

# Iterate through the characters
for i, char in enumerate(char_list):    
    if z <= 1:
        num_spaces = 0
    else:
        # The period of the zigzag is 2 * (z - 1)
        L = 2 * (z - 1)
        pos = i % L
        
        if pos < z:
            # Increasing phase: 0 to z-1
            num_spaces = pos
        else:
            # Decreasing phase: Calculated by mirroring the position
            num_spaces = L - pos
            
    # Print leading spaces followed by the character
    print((' ' * num_spaces) + char)


current_spaces = 0
step = 1

for char in char_list:
    print(' ' * current_spaces + char)
    if z > 1:
        current_spaces += step
        # If we hit the boundaries, reverse the direction
        if current_spaces == z - 1 or current_spaces == 0:
            step *= -1
````

### **Summary of the Zig-Zag Shift logic**

#### **Approach Summary**

The code uses a **Periodic Function** (specifically a Triangle Wave) to determine the indentation.

1. **Cycle Calculation:** It calculates the length of one full "down-and-up" cycle, which is .
2. **Phase Mapping:** It uses the modulo operator `%` to find where the current character sits within that cycle.
3. **Folding Logic:** If the position is in the first half of the cycle, it uses the value directly (increasing). If it's in the second half, it "folds" the value back down to create the decreasing effect.

---

#### **Important Points**

* **The Special Case (z = 1):** If the zigzag width is 1, the period calculation (2 x (z - 1)) would result in zero, causing a "division by zero" error. The code handles this with an `if z <= 1` check to keep spaces at 0.
* **The Zig-Zag Period (L):** To go from  and back to  (for z = 3), we need a cycle of 4 steps (0, 1, 2, 1). The formula L = 2(z - 1) correctly identifies that the pattern repeats every L characters.
* **Mathematical Folding:** * For z = 3, L = 4.
* The sequence of `i % L` is: 0, 1, 2, 3.
* The "folding" math `L - pos` turns that 3 into a 1 (4 - 3 = 1), creating the desired 0, 1, 2, 1 sequence.

#### **Comparison of Phases**

| Step (i) | Modulo (pos) | Condition | Result (num_spaces) |
| --- | --- | --- | --- |
| 0 | 0 | pos < z | 0 |
| 1 | 1 | pos < z | 1 |
| 2 | 2 | pos < z | 2 (Peak) |
| 3 | 3 | pos >= z | 4 - 3 = 1 |
| 4 | 0 | pos < z | 0 (Restart) |

## QUESTION 5.

### Uppercase Every k-th Vowel and lower case other vowels in a File

Write a program that uppercases every k-th vowel in a text file and lowercase other vowels. Vowels are a, e, i, o, u and the check should be case-insensitive. The case of consonants should be unchanged.

Assume k is a positive integer.

The first line of the file contains the integer k.
Subsequent lines of the file contain the text.
The vowel count is cumulative across the entire text. For example, if k=3, the 3rd, 6th, 9th, etc., vowels found in the file should be uppercased, regardless of which line they are on.
**NOTE:** This is a file-in-stdout type question where the input is read from the file and the output is printed in the standard output.

>Example

**Input(contents of the file)**
```text
3
This is a sample text.
it has many vowels.
every third one becomes uppercase.
```
**Output**
```text
This is A sample tExt.
it has mAny vowels.
Every third One becOmes uppErcase.
```

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'uppercase_kth_letter_input.txt'
test_data = """3
This is a sample text.
it has many vowels.
every third one becomes uppercase."""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Code
# ------------------------------------------

with open(filename) as f:
    k = int(f.readline())

    vowel_count = 0
    for line in f:
        for char in line:
            if char.lower() in "aeiou":
                vowel_count += 1
                if vowel_count % k == 0:
                    print(char.upper(),end='')
                else:
                    print(char.lower(),end='')
            else:
                print(char,end='')
````

### Code Summary
The script reads a target integer $k$ from the first line of a file, then scans every subsequent character. It identifies vowels (a, e, i, o, u) regardless of their original case. It maintains a running count of these vowels; if the count is a multiple of $k$, it forces the vowel to uppercase. Otherwise, it forces the vowel to lowercase. All other characters (consonants, numbers, punctuation) remain exactly as they were.

---

### How It Works

#### 1. The Setup (Initialization)
* **`f.readline()`**: The code first extracts the very first line to determine the frequency $k$. Since `readline()` moves the file pointer, the subsequent loop naturally starts at the beginning of the actual text.
* **`vowel_count = 0`**: This variable is initialized outside the loops. This is crucial because it allows the "memory" of how many vowels have been seen to persist across different lines.

#### 2. The Nested Iteration
The code uses a two-tier approach to traverse the data:
1.  **Outer Loop (`for line in f`)**: Iterates through the file line by line. This is memory-efficient for large files.
2.  **Inner Loop (`for char in line`)**: Processes the line character by character.

#### 3. Conditional Logic
For every character, the code asks a series of questions:
* **Is it a vowel?** It checks `char.lower() in "aeiou"`. This handles both 'A' and 'a' correctly.
* **Is it the $k$-th vowel?** If it is a vowel, it increments the counter and uses the **Modulo Operator (`%`)**. 
    * If `vowel_count % k == 0`, the current count is a multiple of $k$ (the 3rd, 6th, 9th, etc.).
* **Transformation**:
    * **$k$-th Vowels**: Forced to `.upper()`.
    * **Other Vowels**: Forced to `.lower()`.
    * **Consonants/Symbols**: Printed via the `else` block with no changes.


### Key Points to Take Care Of

#### 1. Cumulative Counting
The most important aspect is that `vowel_count` is **not** reset at the end of a line. If the 3rd vowel happens to be the last letter of Line 1, the 6th vowel might be in the middle of Line 2. The code handles this correctly by keeping the variable in the outer scope.

#### 2. Case Normalization
The requirement states the check should be case-insensitive, but the output for non-$k$-th vowels must be lowercase. 
> **Note:** If the input text had an uppercase 'I' that was *not* a $k$-th vowel, your code would change it to a lowercase 'i'. This matches your requirement to "lowercase other vowels."

#### 3. Print Formatting
The use of `end=''` in the `print()` function is vital. By default, Python's `print()` adds a newline. Since we are processing the file character-by-character, we need to suppress that newline to keep the original text structure intact.

#### 4. File Pointers
Because the code uses `f.readline()` for $k$, the file pointer is already at line 2 when the `for line in f` loop begins. If you were to call `f.read()` before the loop, the loop would have nothing left to read.

### Simple Stemmer
Given a multiline text file containing lowercase words separated by spaces or newlines, stem each word (remove certain suffixes) according to the list of suffixes given in the suffixes list and print the contents in the same format.

Do not change the line or order of the words.
```text
suffixes = [
    'wards', 'ations', 'ation', 'tions', 'tion', 'asions', 
    'asion', 'sions', 'sion', 'ment', 'ness', 'ship', 
    'hood', 'able', 'ible', 'less', 'ward', 'wise', 'ion', 'ity', 'age',
    'ize', 'ise', 'ify', 'ate', 'ful', 'ous', 'ish', 'ive', 'ing', 'ers', 'er',
    'or', 'ty', 'en', 'ic', 'al', 'ly'
]
```
Remove the first suffix from the list of suffixes that matches a given word. For example for the word "education" the suffix "ation" and "tion" matches but "ation" is present first in the suffixes list thus the word is stemmed as "educ".

**NOTE:** This is a file-in-stdout type question where the input is read from the file and the output is printed in the standard output.

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'simple_stemmer_input.txt'
test_data = """running faster stronger
organization management
happiness joyfulness sadness"""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Code
# ------------------------------------------

suffixes = [
    'wards', 'ations', 'ation', 'tions', 'tion', 'asions', 
    'asion', 'sions', 'sion', 'ment', 'ness', 'ship', 
    'hood', 'able', 'ible', 'less', 'ward', 'wise', 'ion', 'ity', 'age',
    'ize', 'ise', 'ify', 'ate', 'ful', 'ous', 'ish', 'ive', 'ing', 'ers', 'er',
    'or', 'ty', 'en', 'ic', 'al', 'ly'
]


def stem(word: str) -> str:
    """Helper function to stem a word by removing suffixes."""
    for suffix in suffixes:
        n = len(suffix)
        if word.endswith(suffix):
            return word[:-n]
    return word
    
with open(filename) as f:
    for line in f:
        words = line.split()
        print(" ".join(stem(word) for word in words))
````

### How It Works

#### 1. Priority-Based Matching
The `suffixes` list is ordered specifically (generally from longest to most specific). This is crucial because:
* In the word **"education"**, both `ations` and `ation` might look like candidates. 
* Because the code uses a `for suffix in suffixes:` loop with an immediate `return`, it follows a **"first-match-wins"** rule. 
* Once `ation` is found, the function exits, preventing shorter suffixes like `ion` from being processed further on the same word.

#### 2. The `stem(word)` Logic
* **`word.endswith(suffix)`**: A built-in Python string method that returns `True` if the word terminates with that specific string.
* **`word[:-n]`**: This uses **string slicing**. If `n` is the length of the suffix, `[:-n]` tells Python to keep everything from the start of the string up until $n$ characters from the end.

#### 3. Maintaining File Structure
* **`line.split()`**: This breaks a line into a list of words, automatically handling multiple spaces or tabs.
* **`" ".join(...)`**: After the words are stemmed, this puts them back together into a single string separated by single spaces. This ensures that the line-by-line structure of the original file is preserved in the output.


### Key Points to Take Care Of

#### 1. Order Matters (The "Shadowing" Problem)
If you moved a short suffix like `ion` to the very beginning of the list, a word like "education" would become "educat" (removing `ion`) instead of "educ" (removing `ation`). To get the most accurate stem, longer or more complex suffixes should generally appear earlier in the list than the shorter suffixes they contain.

#### 2. Single Pass Only
This specific implementation only performs **one** stemming operation per word. 
> **Example:** In the word "meaningfully", if the code finds `ly` first, it returns "meaningful". It will **not** loop back to see if it can also remove `ful` to get "meaning". If you wanted "recursive" stemming, you would need a `while` loop inside the function.

#### 3. Punctuation Sensitivity
The current `line.split()` approach assumes the "words" are purely alphabetical. If a word in the file is `"running!"`, `word.endswith("ing")` will return `False` because of the exclamation point. In a real-world scenario, you would need to strip punctuation before stemming.

#### 4. Empty Strings and Short Words
The code is safe against short words; if a word is shorter than the suffix, `endswith()` simply returns `False`. However, if a word *is* exactly the suffix (e.g., the word "age"), it will return an empty string.

### String Rearrangement
Write a program that takes a text file with several lines. Each line has a string followed by indices separated by commas. The function should rearrange the characters in the string according to the given indices and print the updated strings.

The indices specify the new order of characters in the string.
Each number corresponds to the position (1-based index) in the original string.
**NOTE:** This is a file-in-stdout type question where the input is read from the file and the output is printed in the standard output.

>Examples

**Input**
```text
hello,2,1,3,5,4
abcdef,3,2,1,6,5,4
xyz,3,1,2
```
**Output**
```text
ehlol
cbafed
zxy
```

````python
# ==========================================
# SETUP: Create input file (Helper code)
# ==========================================
filename = 'string_rearrangement_input.txt'
test_data = """hello,2,1,3,5,4
abcdef,3,2,1,6,5,4
xyz,3,1,2"""

with open(filename, 'w') as f:
    f.write(test_data)

# ------------------------------------------
# SOLUTION: Code
# ------------------------------------------

with open(filename, 'r') as file:
    for line in file:
        # Split the line into the string and indices
        parts = line.strip().split(',')
        original_string = parts[0]
        indices = list(map(int, parts[1:]))

        # Rearrange the string based on indices
        rearranged_string = ''.join(original_string[i - 1] for i in indices)
        print(rearranged_string)
````

### Code Summary
The program reads each line of a file, separates the word from its trailing list of numbers, and treats those numbers as "coordinates." It then constructs a new string by looking up the character at each specified position one by one.

### How It Works

#### 1. Parsing the Input (`split` and `map`)
* **`line.strip().split(',')`**: This handles two tasks at once. `strip()` removes any hidden newline characters (`\n`) at the end of the line, and `split(',')` breaks the line into a list where the first element is the word and the rest are the numeric indices.
* **`map(int, parts[1:])`**: Since the split numbers are initially read as strings (e.g., `"2"`), `map` converts them all into actual integers so they can be used for mathematical indexing.

#### 2. The Rearrangement Logic
The core of the program is this generator expression:
`''.join(original_string[i - 1] for i in indices)`

* **1-Based vs. 0-Based Indexing**: The problem uses **1-based indexing** (where the first letter is 1). However, Python uses **0-based indexing** (where the first letter is 0). The code handles this by subtracting 1 (`i - 1`) from every index provided.
* **The Lookup**: If the input is `hello,2,1...`, the loop first looks for index `2`. It calculates `2 - 1 = 1`. In "hello", index `1` is 'e'. It grabs 'e' and moves to the next number.
* **`''.join(...)`**: This takes all the individual characters found during the lookup and glues them back together into a single string.

### Key Points to Take Care Of

#### 1. Off-by-One Errors
The most common mistake in these problems is forgetting the `i - 1`. If you used the indices directly, a "1" would point to the second letter of the word, and a "5" would likely cause an `IndexError` on a 5-letter word.

#### 2. Data Type Integrity
The indices must be converted to integers. If you try to do `original_string["2"]`, Python will raise a `TypeError` because you cannot index a string using another string.

#### 3. Whitespace and Newlines
Using `.strip()` is vital. Without it, the last index on a line might be read as `"4\n"`. While `int()` is usually smart enough to ignore that whitespace, stripping ensures the data is clean before processing.

#### 4. Index Consistency
The code assumes that the number of indices provided matches the length of the string and that every index is valid (within the range of the string length). If an index is higher than the string length, the program will crash with an `IndexError`.


#### Comparison Table: 1-Based vs. 0-Based Logic
| Original String | "h" | "e" | "l" | "l" | "o" |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **Problem Index (1-based)** | 1 | 2 | 3 | 4 | 5 |
| **Python Index (0-based)** | 0 | 1 | 2 | 3 | 4 |
