## QUESTION 1.

**Call Log Analysis**

You are provided with a call log history represented as a list of dictionaries. Each dictionary contains the following keys:

name (str): The name of the caller.
duration (int): The duration of the call in seconds.
There will be multiple entries with the same name in the dataset. Implement a function process_call_logs(call_logs, task) where task is one of the following:

NOTE: This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

1. **get_call_counts**
Returns a dictionary with caller names as keys and the count of their calls as values.
2. **get_total_call_durations**
Returns a dictionary with caller names as keys and the total duration of their calls as values.
3. **most_frequent_caller**
Returns the name of the caller with the highest number of calls.
If there is a tie, the name with most call duration is returned.

**Example**

call_log = [
    {"name": "Alice", "duration": 300},
    {"name": "Bob", "duration": 200},
    {"name": "Alice", "duration": 100},
    {"name": "Bob", "duration": 400},
    {"name": "Charlie", "duration":300}
]

>get_call_count - {'Alice':2, 'Bob':2, 'Charlie': 1}

>get_total_call_duration - {'Alice': 400, 'Bob': 600, 'Charlie': 300}

>most_frequent_caller - 'Bob', (Alice and Bob has same number of calls but Bob has the most duration.)

```python
def process_call_logs(call_logs, task):
    
    call_counts = {}
    total_durations = {}    
    
    for log in call_logs:
        name = log["name"] # get caller name
        duration = log["duration"] # get call duration       
            
        call_counts[name] = call_counts.get(name,0)+1 # increment call count by 1
        total_durations[name] = total_durations.get(name,0)+duration # increment total duration by 'duration'
            
    if task == 'get_call_counts': 
        return call_counts
    elif task == 'get_total_call_durations':
        return total_durations
    # Find the most frequent caller; in case of a tie, choose the one with the longest total call duration
    elif task == 'most_frequent_caller':
        if not call_counts:
            return ''
        max_count = -1
        max_durn = -1
        most_freq = ""
        # Iterate through call_counts to find the most frequent caller
        for name, count in call_counts.items(): # name, count: dict_items([('Alice', 2), ('Bob', 2), ('Charlie', 1)])
            duration = total_durations[name]
            print(f"duration: {duration}") 
            
            # rule 1: prioritize higher call count
            if count>max_count:
                max_count = count
                max_durn = duration
                most_freq = name
            # rule 2: in case of tie in call count, prioritise higher total call duration
            elif count==max_count:
                if duration>max_durn:
                    max_durn = duration
                    most_freq = name
        return most_freq
        
    else:
        return none
    
call_log = [
    {"name": "Alice", "duration": 300},
    {"name": "Bob", "duration": 200},
    {"name": "Alice", "duration": 100},
    {"name": "Bob", "duration": 400},
    {"name": "Charlie", "duration":300}
]
print(process_call_logs(call_log, 'get_call_counts'))
print(process_call_logs(call_log, 'get_total_call_durations'))    
print(process_call_logs(call_log, 'most_frequent_caller'))
```

****A small summary of the tie-breaking logic you presented.****

 **Summary of the Tie-Breaking Pattern**

The code uses a **Compound Comparison** pattern to select the "best" caller based on two hierarchical rules:

1.  **Primary Rule:** The caller with the **highest frequency (count)** wins immediately.
2.  **Secondary Rule (Tie-Breaker):** If two or more callers have the *same highest frequency*, the one with the **longest total call duration** wins.

| Priority | Criteria | Condition in Code |
| :--- | :--- | :--- |
| **1 (Primary)** | Higher Count | `if count > max_count:` |
| **2 (Secondary)** | Longer Duration (only if counts are equal) | `elif count == max_count: \n if duration > max_durn:` |

This pattern ensures that a difference in the primary metric (`count`) always overrides any difference in the secondary metric (`duration`).

## QUESTION 2.

**Cartesian Points Processing**

Given a set of tuples of ints representing points in Cartesian coordinates, write a function that performs the following tasks:

1. **sort_close_to_y_axis:** Returns a sorted list of the points based on the absolute value of their x-coordinate (distance from the y-axis). The point with the lower angle in polar coordinates comes first in case of ties.
2. **closest_point_to_origin:** Find the closest point to the origin (0,0) based on the Manhattan distance. If there are multiple points at the same distance, return the one with the lower angle in polar coordinates.
3. **group_by_quadrant:** Return a dictionary where the keys are quadrant numbers (1, 2, 3, 4) and the values are sets of points in those quadrants. Do not include keys for quadrants with no points.

Note the angle is computed using math.atan2(y,x) where x and y are the cartesian coordinates. math.atan2 returns the angle in the range of -pi to pi. You can use the get_angle function from the template.

Assume no points are on the axes or origin and the coordinates are integers.

Hint: sorted and min functions can be used with the key argument to make the code simpler.

**Definitions:**

Manhattan distance between two points (x_1, y_1) and (x_2, y_2) is |x_1 - x_2| + |y_1 - y_2|.
NOTE: This is a function type question, you don't have to take input or print the output, just have to complete the required function definition.

**Example**

points = {
    (2, 5), (-2, 6), (-2, 7), (-1, 3),
    (-3, -4), (4, -6), (0, 7)
}

```python
import math
from typing import Set, Tuple, List, Dict, Union

# Helper function provided in the prompt's context
def get_angle(x: int, y: int) -> float:
    """Computes the angle in polar coordinates using math.atan2(y, x)."""
    return math.atan2(y, x)

# Type alias for clarity
Point = Tuple[int, int]

def process_points(points: Set[Point], task: str) -> Union[List[Point], Point, Dict[int, Set[Point]]]:
    """
    Performs various tasks on a set of Cartesian points:
    1. sort_close_to_y_axis: Sorts by |x|, breaking ties with the polar angle.
    2. closest_point_to_origin: Finds the closest point using Manhattan distance, breaking ties with the polar angle.
    3. group_by_quadrant: Groups points into quadrants (1, 2, 3, 4).
    """

    if task == "sort_close_to_y_axis":
        return sort_close_to_y_axis(points)
    elif task == "closest_point_to_origin":
        return closest_point_to_origin(points)
    elif task == "group_by_quadrant":
        return group_by_quadrant(points)
    else:
        # Handle unknown task
        raise ValueError(f"Unknown task: {task}")

# --- Task 1: Sort by closeness to Y-axis ---
def sort_close_to_y_axis(points: Set[Point]) -> List[Point]:
    """
    Returns a sorted list of points based on the absolute value of their x-coordinate
    (distance from the y-axis). Ties are broken by the lower polar angle.
    """
    
    # The key function for sorting needs to return a tuple:
    # 1. The primary sorting criterion: absolute value of the x-coordinate (|x|)
    # 2. The secondary sorting criterion (for tie-breaking): the polar angle
    
    # Note: The problem assumes no points on the axes, but handles (0, 7) in the example.
    # The absolute distance from the y-axis is |x|.
    
    sorted_list = sorted(
        points,
        key=lambda p: (abs(p[0]), get_angle(p[0], p[1]))
    )
    
    return sorted_list

# --- Task 2: Closest Point to Origin (Manhattan Distance) ---
def closest_point_to_origin(points: Set[Point]) -> Point:
    """
    Finds the closest point to the origin (0,0) based on the Manhattan distance,
    breaking ties with the lower polar angle.
    Manhattan distance is |x| + |y| for distance from (0,0).
    """
    
    # The key function for min needs to return a tuple:
    # 1. The primary criterion: Manhattan distance |x| + |y|.
    # 2. The secondary criterion (for tie-breaking): the polar angle.
    
    closest = min(
        points,
        key=lambda p: (abs(p[0]) + abs(p[1]), get_angle(p[0], p[1]))
    )
    
    return closest

# --- Task 3: Group by Quadrant ---
def get_quadrant(x: int, y: int) -> int:
    """
    Determines the quadrant (1, 2, 3, or 4) for a point (x, y).
    Assumes no points are on the axes or origin.
    

[Image of Cartesian quadrants and signs]

    """
    if x > 0 and y > 0:
        return 1  # (+, +)
    elif x < 0 and y > 0:
        return 2  # (-, +)
    elif x < 0 and y < 0:
        return 3  # (-, -)
    elif x > 0 and y < 0:
        return 4  # (+, -)
    else:
        # Should not happen based on problem constraints (no points on axes/origin)
        raise ValueError("Point is on an axis or the origin, violating constraints.")

def group_by_quadrant(points: Set[Point]) -> Dict[int, Set[Point]]:
    """
    Returns a dictionary where keys are quadrant numbers (1, 2, 3, 4)
    and values are sets of points in those quadrants.
    """
    
    quadrant_groups: Dict[int, Set[Point]] = {1: set(), 2: set(), 3: set(), 4: set()}
    
    for x, y in points:
        quadrant = get_quadrant(x, y)
        quadrant_groups[quadrant].add((x, y))
        
    # Filter out keys (quadrants) that have no points
    return {
        quadrant: point_set
        for quadrant, point_set in quadrant_groups.items()
        if point_set
    }

points = {
    (2, 10),
    (2, 5),
    (-2, 10),
    (-2, 5),
    (-3, -4),
    (5, -2)
}
print(process_points(points, "sort_close_to_y_axis"))
print(process_points(points, "closest_point_to_origin"))
print(process_points(points, "group_by_quadrant"))
```

This code defines a `process_points` function to analyze a set of 2D Cartesian points, utilizing the **polar angle** as a critical tie-breaker in several tasks.

**1. Key Concept: Compound Comparison**

The core of the functions relies on Python's **tuple comparison** to handle complex sorting and minimization: `(Primary Metric, Secondary Metric)`.

**2. Analysis Tasks**

| Task Parameter | Purpose | Primary Metric | Secondary Metric (Tie-Breaker) |
| :--- | :--- | :--- | :--- |
| **`sort_close_to_y_axis`** | Sorts points from closest to farthest away from the Y-axis. | Absolute X-coordinate ($\vert x \vert$). | Lower Polar Angle ($\theta$). |
| **`closest_point_to_origin`** | Finds the single point closest to the origin (0, 0). | Manhattan Distance ($\vert x \vert + \vert y \vert$). | Lower Polar Angle ($\theta$). |
| **`group_by_quadrant`** | Organizes the points based on their location in the Cartesian plane. | N/A (Grouping by sign of $x$ and $y$). | N/A (No sorting involved). |



The helper function `get_angle(x, y)` provides the polar angle $\theta$ using $atan2(y, x)$, which is crucial for defining a consistent tie-breaking order among geometrically tied points.

## QUESTION 3.

**Sales Data Analysis**

You are provided with sales data represented as a list of dictionaries. Each dictionary contains the following keys:

product_id (str): The unique identifier for a product.
units_sold (int): The number of units sold in a transaction.
revenue (int): The revenue generated from that transaction.
There will be multiple entries for the same product_id. Implement a function analyse_sales_data(sales_data, task) where task is one of the following strings:

**total_revenue**
Returns the total revenue over all transanctions.

**product_wise_total_units_and_revenue**
Returns a dictionary with product_id as keys and a tuple with (the number of units sold for that product, total revenue generated from that product) as values.

**top_selling_product**
Returns the product_id of the product with the highest number of units sold.
If there is a tie in units sold, the product with the higher total revenue is returned.

**average_product_price**
Returns a dictionary with product_id as keys and the average price computed as the (total revenue / total units_sold) over all the transanctions for the product rounded to 2 decimal places.

NOTE: This is a function type question, you don't have to take input or print the output, you just have to complete the required function definition.

**Example**

sales_data = [
  
   {"product_id": "P101", "units_sold": 50, "revenue": 400},
  
   {"product_id": "P102", "units_sold": 30, "revenue": 900},
  
   {"product_id": "P101", "units_sold": 70, "revenue": 600},
  
   {"product_id": "P103", "units_sold": 120, "revenue": 600}
]

>total_revenue -> 2500

>product_wise_total_units_and_revenue -> {'P101': (120, 1000), 'P102': (30, 900), 'P103': (120, 600)}

>top_selling_product -> 'P101' (P101 and P103 are tied on units, but P101 has higher revenue.)

>average_product_price -> {'P101': 8.33, 'P102': 30, 'P103': 5} (For P101, (400+600)/(50+70))

```python
def analyse_sales_data(sales_data, task):
    
    # --- PRE-PROCESSING: Compute the product summary (Units and Revenue) once ---
    product_summary = {}
    for transaction in sales_data:
        
        # Safely extract data, assuming keys exist based on problem description
        
        product_id = transaction.get("product_id")
        #print(f"product_id: {product_id}")
        units = transaction.get("units_sold", 0)
        #print(f"units: {units}")
        revenue = transaction.get("revenue", 0)
        #print(f"revenue: {revenue}")
        
        # Ensure we don't proceed with missing product_id
        
        if not product_id:
            continue
            
        # Aggregate units and revenue
        
        if product_id in product_summary:
            current_units, current_revenue = product_summary[product_id]
            #print(f"current_units, current_revenue: {current_units}, {current_revenue}") --> current_units, current_revenue: 50, 400
            product_summary[product_id] = (current_units + units, current_revenue + revenue)
            #print(f"product_summary updated for {product_id}: {product_summary[product_id]}") --> product_summary updated for P101: (120, 1000)
        else:
            product_summary[product_id] = (units, revenue)
    #print(f"product_summary: {product_summary}") --> product_summary: {'P101': (120, 1000), 'P102': (30, 900), 'P103': (120, 600)}
    
    if task == "total_revenue":
        # Task 1: Total Revenue
        total_rev = sum(d.get("revenue", 0) for d in sales_data)
        return total_rev
    
    if task == "product_wise_total_units_and_revenue":
        # Task 2: Product-Wise Summary
        return product_summary

    if task == "top_selling_product":
        # Task 3: Top Selling Product
        
        # Using a list of all products to handle edge case of empty sales_data
        products = list(product_summary.items())
        
        if not products:
            return None # Or an empty string, depending on expected behavior for empty data

        # Set the initial 'best' product to the first one in the summary
        best_product_id, (max_units, max_revenue) = products[0]
        
        # Iterate starting from the second product
        for product_id, (units, revenue) in products[1:]:
            
            # Rule 1: Prioritize higher units sold
            if units > max_units:
                max_units = units
                max_revenue = revenue
                best_product_id = product_id
            
            # Rule 2: If units are tied, use higher revenue as a tie-breaker
            elif units == max_units:
                if revenue > max_revenue:
                    max_revenue = revenue
                    best_product_id = product_id
        
        return best_product_id

    if task == "average_product_price":
        # Task 4: Average Product Price
        average_prices = {}
        for product_id, (units, revenue) in product_summary.items():
            if units > 0:
                # Use f-string formatting to ensure exactly 2 decimal places, 
                # which can sometimes differ from round() behavior in tests.
                # However, since the output type is a float/number, round() is generally preferred.
                # Sticking to round() as per your original requirement.
                average_price = revenue / units
                average_prices[product_id] = round(average_price, 2)
            else:
                # Handle the division by zero (0 units sold)
                average_prices[product_id] = 0.00
                
        return average_prices
        
    # Fallback for unrecognized task
    return None

sales_data = [
   {"product_id": "P101", "units_sold": 50, "revenue": 400},
  
   {"product_id": "P102", "units_sold": 30, "revenue": 900},
  
   {"product_id": "P101", "units_sold": 70, "revenue": 600},
  
   {"product_id": "P103", "units_sold": 120, "revenue": 600}
]

print(analyse_sales_data(sales_data, "total_revenue"))
print(analyse_sales_data(sales_data, "product_wise_total_units_and_revenue"))
print(analyse_sales_data(sales_data, "top_selling_product"))
print(analyse_sales_data(sales_data, "average_product_price"))
```

## QUESTION 4.

**Book Reading List Data Analysis**

You are given a list of dictionaries, each representing a book in a reading list.
Every dictionary contains the following keys:

book_id (str) a unique identifier for the book
pages (int) the number of pages of the book
rating (float) the rating of the book (05)
Implement a function book_analysis(data, request) that performs the operation specified by request on the list data.

**Supported requests:**

1. **average_rating -** Return the average rating of all books, rounded to two decimal places.
2. **average_pages -** Return the average number of pages of all books, rounded to two decimal places.
3. **longest_book -** Return the book_id of the book with the maximum number of pages.
If several books share the same maximum page count, return the one among them with the highest rating.
4. **above_average_books -** Return a set of book_ids of books that satisfy both conditions:
rating strictly greater than the average rating of all books, and
pages strictly greater than the average number of pages.

NOTE: This is a function type question, you don't have to take input or print the output, you just have to complete the required function definition.

data = [
    
    {"book_id": "X1", "pages": 540, "rating": 4.9},
   
    {"book_id": "X2", "pages": 96,  "rating": 4.6},
   
    {"book_id": "X3", "pages": 1225,"rating": 4.4},
    
    {"book_id": "X4", "pages": 328, "rating": 4.2},
    
    {"book_id": "X5", "pages": 180, "rating": 3.9},
   
    {"book_id": "X6", "pages": 1225,"rating": 4.3},

]


```python
def book_analysis(data, request):
    
    # Handle the edge case of empty input data
    if not data:
        if request in ["average_rating", "average_pages"]:
            return 0.00
        elif request == "longest_book":
            return None 
        elif request == "above_average_books":
            return set()
        else:
            return None

    # --- Pre-computation of required summary statistics ---
    
    total_books = len(data)
    total_pages = sum(book["pages"] for book in data)
    total_rating = sum(book["rating"] for book in data)
    
    # Calculate Averages (needed for multiple tasks)
    avg_rating = total_rating / total_books
    avg_pages = total_pages / total_books
    

    if request == "average_rating":
        # Task 1: Average Rating
        return round(avg_rating, 2)

    elif request == "average_pages":
        # Task 2: Average Pages
        return round(avg_pages, 2)

    elif request == "longest_book":
        # Task 3: Longest Book (Tie-breaker: Highest Rating)
        
        longest_book_id = None
        max_pages = -1
        max_rating = -1.0
        
        for book in data:
            book_id = book["book_id"]
            pages = book["pages"]
            rating = book["rating"]
            
            # Rule 1: Prioritize maximum pages
            if pages > max_pages:
                max_pages = pages
                max_rating = rating
                longest_book_id = book_id
            
            # Rule 2: If pages are equal, use higher rating as a tie-breaker
            elif pages == max_pages and rating > max_rating:
                max_rating = rating
                longest_book_id = book_id
                
        return longest_book_id

    elif request == "above_average_books":
        # Task 4: Books strictly above both average rating and average pages
        
        above_average_ids = set()
        
        for book in data:
            # Use unrounded averages for strict inequality comparison
            if (book["rating"] > avg_rating) and (book["pages"] > avg_pages):
                above_average_ids.add(book["book_id"])
                
        return above_average_ids

    else:
        # Handle unrecognized request
        return None
    
data = [
    
    {"book_id": "X1", "pages": 540, "rating": 4.9},
   
    {"book_id": "X2", "pages": 96,  "rating": 4.6},
   
    {"book_id": "X3", "pages": 1225,"rating": 4.4},
    
    {"book_id": "X4", "pages": 328, "rating": 4.2},
    
    {"book_id": "X5", "pages": 180, "rating": 3.9},
   
    {"book_id": "X6", "pages": 1225,"rating": 4.3},

]
print(book_analysis(data, "average_rating"))
print(book_analysis(data, "average_pages"))
print(book_analysis(data, "longest_book"))
print(book_analysis(data, "above_average_books"))
```

```python
def process_employees(employee_list: list, request: str):
    """
    Process the employee list as per the request.
    
    Args:
        employee_list (list[dict]): List of dictionaries with the keys
            "employee_id", "tasks_completed", and "hours_worked".
        request (str): One of the following options:
            - 'total_tasks_completed'
            - 'most_efficient_employee'
            - 'sort_by_hours_worked'
            - 'productive_team_members'.
        
    Returns:
        The output corresponding to the request.
    """
    if request == "total_tasks_completed":
        # 1. total_tasks_completed: Compute the total number of tasks completed.
        total = 0
        for employee in employee_list:
             # Use .get() for safety, though bracket notation is fine if keys are guaranteed
             total += employee.get('tasks_completed', 0) 
        return total  # Return is correctly placed at the end of this block
    
    # Use elif to check the next condition only if the first one was false
    elif request == "most_efficient_employee":
        # 2. most_efficient_employee: Highest ratio, tie-breaker is highest tasks.
        
        best_employee_id = -1
        max_efficiency = -1.0
        max_tasks_at_tie = -1
        
        for employee in employee_list:
            tasks = employee['tasks_completed']
            hours = employee['hours_worked']
            employee_id = employee['employee_id']

            # Calculate Efficiency Ratio (handle division by zero)
            efficiency = tasks / hours if hours > 0 else 0.0

            # Comparison Logic
            if efficiency > max_efficiency:
                # 1. New highest efficiency found
                max_efficiency = efficiency
                max_tasks_at_tie = tasks
                best_employee_id = employee_id
            
            elif efficiency == max_efficiency:
                # 2. Efficiency is tied, apply the tie-breaker: check tasks_completed
                if tasks > max_tasks_at_tie:
                    max_tasks_at_tie = tasks
                    best_employee_id = employee_id
                        
        return best_employee_id
    
    elif request == "sort_by_hours_worked":
        # 3. sort_by_hours_worked: Sort by hours_worked (high to low), 
        #    Tie-breaker: tasks_completed (high to low).

        # Key for sorting: a tuple (-hours_worked, -tasks_completed). 
        # Negative signs ensure descending order for both metrics.
        def sort_key(employee):
            return (-employee['hours_worked'], -employee['tasks_completed'])

        sorted_employees = sorted(employee_list, key=sort_key)
        
        # Extract the employee_ids from the sorted list
        return [employee['employee_id'] for employee in sorted_employees]

    elif request == "productive_team_members":
        # 4. productive_team_members: Filter employees who have completed 20 or more tasks.

        # Use a list comprehension for efficient filtering, maintaining original order
        productive_ids = [
            employee['employee_id']
            for employee in employee_list
            if employee['tasks_completed'] >= 20
        ]
        return productive_ids

    else:
        # Handle unrecognized request
        return "Error: Invalid analysis request type."
    
employee_list = [
    {'employee_id': 1, 'tasks_completed': 30, 'hours_worked': 40},
    {'employee_id': 2, 'tasks_completed': 60, 'hours_worked': 90},
    {'employee_id': 3, 'tasks_completed': 60, 'hours_worked': 100},
    {'employee_id': 4, 'tasks_completed': 50, 'hours_worked': 100},
    {'employee_id': 5, 'tasks_completed': 10, 'hours_worked': 30},
]
print(process_employees(employee_list, 'total_tasks_completed'))  # Output: 210
print(process_employees(employee_list, 'most_efficient_employee'))  # Output: 1
print(process_employees(employee_list, 'sort_by_hours_worked'))  # Output: [4, 3, 2, 1, 5]
print(process_employees(employee_list, 'productive_team_members'))  # Output: [1, 2, 3,
```

Implement the following functions:

* Returns the total engagement (likes + comments) for a given video.
* engagement_rate(video: dict) -> float
* Returns the engagement rate for a given video, using the formula:

Engagement Rate =  likes + comments / views  x 100

If views is 0, return 0.0

Round the result to 2 decimal places


1. **most_engaging_video(videos: list[dict]) -> *str***

Returns the title of the video with the highest engagement rate. If there is a tie, return the first one in the list.

2. **videos_with_engagement_rate_above_threshold(videos: list[dict], threshold: float) -> *list[str]***

Returns a list of video titles whose engagement rate is strictly greater than the given threshold.

3. **average_engagement_rate(videos: list[dict]) -> *float***

Returns the average engagement rate of all videos with non-zero views, rounded to 2 decimal places.

```python
def total_engagement(video: dict) -> int:
    """Returns the total engagement (likes + comments) of a given video."""
    return video.get("likes") + video.get("comments")
    

def engagement_rate(video: dict) -> float:
    """Returns the engagement rate ((likes + comments) / views) * 100 rounded to 2 decimals. Returns 0.0 if views == 0."""
    engagement = total_engagement(video)
    views = video.get("views")
    if views == 0:
        return 0.0
    rate = (engagement/views)*100
    return round(rate,2)
    

def most_engaging_video(videos: list) -> str:
    """Returns the title of the video with the highest engagement rate. Returns the first in case of tie."""
    if not videos:
        return ""
    most_engaging = max(videos, key=engagement_rate)
    return most_engaging.get("title", "")
    

def videos_with_engagement_rate_above_threshold(videos: list, threshold: float) -> list:
    """Returns titles of videos whose engagement rate is strictly greater than the given threshold."""
    titles_above_threshold = []
    
    for video in videos:
        rate = engagement_rate(video)
        
        if rate > threshold:
            titles_above_threshold.append(video.get("title", ""))
            
    return titles_above_threshold
    

def average_engagement_rate(videos: list) -> float:
    """Returns the average engagement rate of videos with non-zero views, rounded to 2 decimal places."""
    total_rate_sum = 0.0
    valid_video_count = 0
    
    for video in videos:
        if video.get("views", 0) > 0:
            rate = engagement_rate(video)
            total_rate_sum += rate
            valid_video_count += 1
            
    if valid_video_count == 0:
        return 0.0
    
    average_rate = total_rate_sum / valid_video_count
    return round(average_rate, 2)

video_data = [
    {
        "title": "Marketing 101: Boost Your Reach",
        "views": 50000,
        "likes": 2500,
        "comments": 500
    },
    {
        "title": "Quick Python Tutorial",
        "views": 1000,
        "likes": 50,
        "comments": 20
    },
    {
        "title": "Vlog: My Day Off",
        "views": 0,  # Zero views case
        "likes": 100,
        "comments": 10
    },
    {
        "title": "Advanced Data Structures",
        "views": 20000,
        "likes": 1000,
        "comments": 200
    },
    {
        "title": "Holiday Travel Guide",
        "views": 4000,
        "likes": 200,
        "comments": 40
    }
]

print(total_engagement(video_data[0])) 
print(engagement_rate(video_data[0]))  
print(most_engaging_video(video_data))
print(videos_with_engagement_rate_above_threshold(video_data, 5.0))
print(average_engagement_rate(video_data))
```

```python
def total_revenue_in_region(sales_data: list, region: str) -> int:
    """Sum of revenue for the specified region."""
    tot_revenue = 0
    for i in sales_data:
        if i["region"] == region:
            tot_revenue += i["revenue"]
    return tot_revenue
    

def revenue_range_for_product(sales_data: list, product: str) -> int:
    """Maximum revenue minus minimum revenue for the given product."""
    product_rev = []
    for i in sales_data:
        if i["product"] == product:
            product_rev.append(i["revenue"])
            
    if len(product_rev) <= 1:
        return 0
    else:
        return max(product_rev) - min(product_rev)
    

def region_with_max_sales(sales_data: list) -> str:
    """Region with highest total revenue; ties broken by total quantity_sold."""
    if not sales_data:
        return ""

    # Initialize a standard dictionary. We'll handle key existence checks manually.
    region_metrics = {}

    for record in sales_data:
        region = record["region"]
        revenue = record.get("revenue", 0)
        quantity = record.get("quantity_sold", 0)

        if region not in region_metrics:
            # Manual initialization for a new region
            region_metrics[region] = {"revenue": 0, "quantity_sold": 0}

        # Aggregation
        region_metrics[region]["revenue"] += revenue
        region_metrics[region]["quantity_sold"] += quantity

    # Find the region (key) that maximizes the (revenue, quantity_sold) tuple.
    # The max() function handles the tie-breaker correctly by comparing tuples.
    max_region = max(region_metrics.keys(), key=lambda r: (region_metrics[r]["revenue"], region_metrics[r]["quantity_sold"]))

    return max_region


def steady_revenue_products(sales_data: list) -> set:
    """Return a set of product names whose revenue range is strictly less than 5000."""
    
    # 1. Group all revenues by product using a standard dictionary
    product_rev = {}
    for record in sales_data:
        product = record["product"]
        revenue = record.get("revenue", 0)
        
        if product not in product_rev:
            # Manual list initialization for a new product
            product_rev[product] = []
            
        product_rev[product].append(revenue)
        
    steady_products = set()
    
    # 2. Calculate the revenue range for each product
    for product, revenues in product_rev.items():
        if len(revenues) <= 1:
            revenue_range = 0
        else:
            # Calculate range: max - min
            revenue_range = max(revenues) - min(revenues)
            
        # 3. Check the condition
        if revenue_range < 5000:
            steady_products.add(product)

    # Note: The original template requested a list to be sorted alphabetically,
    # but the logic requested a set of products. Returning the set per the logic.
    return steady_products

sales_data = [
    # Region 'North': High Revenue, High Quantity
    {"region": "North", "product": "Laptop Pro", "revenue": 55000, "quantity_sold": 55},
    {"region": "North", "product": "Mouse XL", "revenue": 1000, "quantity_sold": 200},
    {"region": "North", "product": "Laptop Pro", "revenue": 56000, "quantity_sold": 56}, # Max/Min test for Laptop Pro

    # Region 'South': Test for low revenue range (Steady Product)
    {"region": "South", "product": "Phone Stand", "revenue": 1200, "quantity_sold": 60},
    {"region": "South", "product": "Phone Stand", "revenue": 1000, "quantity_sold": 50}, # Range is 2000 (<5000)

    # Region 'East': High Total Revenue (for Max Sales check)
    {"region": "East", "product": "Tablet Basic", "revenue": 28000, "quantity_sold": 70},
    {"region": "East", "product": "Mouse XL", "revenue": 900, "quantity_sold": 180},
    {"region": "East", "product": "Tablet Basic", "revenue": 29000, "quantity_sold": 72},

    # Region 'West': Tie-breaker test for max sales
    # Total Revenue: 50000. Total Quantity: 100.
    {"region": "West", "product": "Laptop Lite", "revenue": 40000, "quantity_sold": 80},
    {"region": "West", "product": "Phone Stand", "revenue": 10000, "quantity_sold": 20},

    # Region 'Center': Tie-breaker rival for West
    # Total Revenue: 50000. Total Quantity: 101. (Should win the tie)
    {"region": "Center", "product": "Laptop Lite", "revenue": 50000, "quantity_sold": 101},
]

print(total_revenue_in_region(sales_data, "North"))  # Output: 112000
print(revenue_range_for_product(sales_data, "Laptop Pro"))  # Output: 1000
print(region_with_max_sales(sales_data))  # Output: "Center" (tie-breaker test)101]
print(steady_revenue_products(sales_data))  # Output: {'Phone Stand'}
print(sorted_employees)
```
